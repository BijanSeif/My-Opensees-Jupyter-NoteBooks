```python
import topspy.timehistories.Random as bTH
import topspy.dynamic as bjD
import matplotlib.pyplot as plt
%matplotlib notebook
```

### Loading some accelerograms

in this part we load some accelerograms and draw their response spectrum

```python
plt.figure(0)
name=[]
for rcd in bTH.Recordslist():
    [dt,acc]=bTH.LAT2(rcd)
    t=[i*dt for i in range(len(acc))]
    plt.plot(t,acc)
    name.append(rcd)
print(name)
```

### Draw Response Spectrums

Now we draw 3 types of response spectrums for previouse accelerations

```python
T=[0.0001+i/10 for i in range(20)]

plt.figure(1)
plt.figure(2)
plt.figure(3)

for rcd in bTH.Recordslist():
    [dt,acc]=bTH.LAT2(rcd)
    [T,Sd,Sv,Sa]=bjD.RspSpc(acc,dt,T,0.05)
    plt.figure(1)
    plt.title('Displacement Spectrum')
    plt.plot(T,Sd)
    plt.figure(2)
    plt.title('Psuedo Velocity Spectrum')
    plt.plot(T,Sv)
    plt.figure(3)
    plt.title('Psuedo Acceleration Spectrum')
    plt.plot(T,Sa)
    
```

### Work more on Elcentro

```python
acc=[i for i in bTH.Recordslist() if i[0:2]=='El'][0]
[dt,acc]=bTH.LAT2(acc)

#Ploting Elcentro time history
plt.figure(4)
t=[i*dt for i in range(len(acc)) ]
plt.plot(t,acc)
plt.title('ElCentro Acceleration time history')

#Plot Response spectrum for different Damping ratio
plt.figure(5)
plt.title('ElCentro Psuedoi Acceleration Response Spectrum')
dampratio=[0, 0.01, 0.02, 0.05, 0.1]
T=[0.0001+i/10 for i in range(20)]
for damp in dampratio:
    [Tl,Sd,Sv,Sa]=bjD.RspSpc(acc,dt,T,damp)
    plt.plot(Tl,Sa,label=f'Damp={damp}')
plt.legend()    

```
